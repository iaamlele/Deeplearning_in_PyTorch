# 搭建人工神经网络

```python
import torch
import torch.optim as optim

#定义变量
#x归一化 x在0.02,0.04..1之间
x = torch.FloatTensor(np.arange(len(counts), dtype = float) / len(counts))
y = torch.FloatTensor(np.array(counts, dtype = float))

sz = 10 #隐含神经元数量
#torch.rand()生成均匀分布的随机数，其中所有的数都在 [0, 1) 的区间内
#torch.randn()生成标准正态分布（均值为0，标准差为1）的随机数，也被称为标准正态随机数
weights = torch.randn((1, sz), requires_grad = True)
biases = torch.randn((sz), requires_grad = True)
weights2 = torch.randn((sz, 1), requires_grad = True)

#迭代训练神经网络
learning_rate = 0.001
losses = []
x = x.view(50, -1) #将张量 x 重塑为一个新的形状，其中第一个维度是 50，而第二个维度根据剩余元素的数量自动计算，总数不变
y = y.view(50, -1)
for i in range(100000):
    hidden = x * weights + biases
    hidden = torch.sigmoid(hidden)
    predictions = hidden.mm(weights2)
    loss = torch.mean((predictions - y) ** 2)
    losses.append(loss.data.numpy())
    
    if i % 10000 == 0:
        print('loss:', loss)
    
    #梯度下降
    loss.backward()
    weights.data.add_(-learning_rate * weights.grad.data)
    biases.data.add_(-learning_rate * biases.grad.data)
    weights2.data.add_(-learning_rate * weights2.grad.data)
    
    weights.grad.data.zero_()
    biases.grad.data.zero_()
    weights2.grad.data.zero_()
    
plt.plot(losses)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```
